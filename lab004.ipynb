```python
import pandas as pd
data=pd.read_csv('./Find_S.csv')
```

```python
print(data)
```

```python
def fs(data):
    attributes=data.iloc[:,:-1].values
    target=data.iloc[:,-1].values
    hyp=None
    for i in range(len(target)):
        if target[i].lower()=="yes":
            hyp=attributes[i].copy()
            break
    if hyp is None:
        raise ValueError("NO POSITIVE EXAMPLES FOUND IN THE DATASET.")

    for i in range(len(target)):
        if target[i].lower()=="yes":
            for j in range(len(hyp)):
                if hyp[j] != attributes[i][j]:
                    hyp[j]='?'
    return hyp
```

```python
final_h=fs(data)
print("MOST SPECIFFIC HYPOTHESIS:",final_h)
```
