```python
numerical_features=housing_df.select_dtypes(include =[np.number]).columns
print(numerical_features)
```

```python
pit.figure(figsize=(15,20))
for i, feature in enumerate(numerical_features):
    pit.subplot(3,3,i+1)
    sns.histplot(housing_df[feature],kde=True, bins=10,color='orange')
    pit.title(f'dDISTRIBUTION OF {feature}')
pit.tight_layout()
```

```python
pit.figure(figsize=(15,20))
for i, feature in enumerate(numerical_features):
    pit.subplot(3,3,i+1)
    sns.boxplot(housing_df[feature],color='orange')
    pit.title(f'BOX PLOT OF {feature}')
pit.tight_layout()
pit.show()
```

```python
print("OUTLINERS DETECTION:")
ol_sum={}
for feature in numerical_features:
    Q1=housing_df[feature].quantile(0.25)
    Q3=housing_df[feature].quantile(0.75)
    IQR=Q3-Q1
    lowb=Q1-1.5*IQR
    uppb=Q3+1.5*IQR
    ol=housing_df[(housing_df[feature]<lowb) | (housing_df[feature]>uppb)]
    ol_sum[feature]=len(ol)
    print(f"{feature}:\t {len(ol)} OUTLINERS")
```
